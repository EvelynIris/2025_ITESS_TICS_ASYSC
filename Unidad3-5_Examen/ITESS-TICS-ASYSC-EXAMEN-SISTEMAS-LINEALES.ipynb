```python
import numpy as np
import matplotlib.pyplot as plt
from thinkdsp import CosSignal
from thinkdsp import Wave

# Creamos las variables con valores comerciales
R = 200     
L = 1        
C = 0.1     
Vm = 1       # voltaje máximo
f = 10       
w = 2*np.pi*f

Z_re = R
Z_im = w*L - 1/(w*C)   # parte imaginaria

Z_mag = np.sqrt(Z_re**2 + Z_im**2)   # |Z|
teta = np.arctan(Z_im / Z_re)       # fase de Z

A = Vm / Z_mag   # amplitud de la corriente
```

```python

signal_i = CosSignal(freq=f, amp=A, offset=-teta)
wave_i = signal_i.make_wave(duration=1, framerate=1000)
```

# Grafica en el tiempo

```python
from thinkdsp import decorate
import matplotlib.pyplot as plt

plt.figure()
wave_i.plot()
plt.title("Respuesta en el tiempo: i(t)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.show()
```

# Espectro

```python
from thinkdsp import decorate
import matplotlib.pyplot as plt
spectrum = wave_i.make_spectrum()

plt.figure()
spectrum.plot()
plt.title("Espectro de magnitud de i(t)")
plt.xlabel("Frecuencia (Hz)")
plt.show()
```
